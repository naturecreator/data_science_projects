```python
import pandas as pd

# Import seaborn, a Python graphing library
import seaborn as sns
import matplotlib.pyplot as plt

# Current version of seaborn generates a collection of warnings that can be ignored
import warnings
warnings.filterwarnings("ignore")

# Load the dataset, and now the dataset becomes a Pandas DataFrame
iris_flower = pd.read_csv("Iris.csv")

# Use head() to display the first 5 rows of the dataset 
iris_flower.head()
```

```python
# Check the number of varities of species present in the dataset
iris_flower["Species"].value_counts()
```

```python
# Let's plot the features of Iris flower using scatter plot
iris_flower.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python
# We can also use seaborn's jointplot to show scatter plot as well as histogram at the same time 
sns.jointplot(data=iris_flower, x="SepalLengthCm", y="SepalWidthCm", size=5)

plt.show()
```

```python
# Use seabborn's FacetGrid to color the scatter plot by species
sns.FacetGrid(iris_flower, hue="Species", size=5)\
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm")\
   .add_legend()

plt.show()
```

```python
sns.boxplot(data=iris_flower, x="Species", y="PetalLengthCm")

plt.show()
```

```python
# We can extend the above mentioned plot by adding a layer of individual points on top of it
# through seaborn's strip plot

#

# Use jitter=True to see easier distribution along vertical lines above species

#

# Axes as ax causes the resulting plot to be shown on top of the previous axes
ax = sns.boxplot(data=iris_flower, x="Species", y="PetalLengthCm")
ax = sns.stripplot(data=iris_flower, x="Species", y="PetalLengthCm", jitter=True, edgecolor="gray")
```

```python
# Violin plot combines the benefits of the aforementioned plots 

sns.violinplot(data=iris_flower, x="Species", y="PetalLengthCm")

plt.show()
```

```python
#kdeplot can be used to fit univariate or bivariate kernel density estimate
sns.FacetGrid(iris_flower, hue="Species", size=5)\
   .map(sns.kdeplot, "PetalLengthCm")\
   .add_legend()

plt.show()
```

```python
# pairplot is used to plot a pair wise relationship in a dataset

# Here, Iris-setosa sepcies is separated from the other two across all feature combinations
sns.pairplot(iris_flower.drop("Id", axis=1), hue="Species", size=2)
```

```python
# In pairplot we can also update diagonal elements, by default it shows "kde"

sns.pairplot(iris_flower.drop("Id", axis=1), hue="Species", size=3, diag_kind="hist")
```

```python
# Used Pandas boxplot on each feature split out by species 
iris_flower.drop("Id", axis=1).boxplot(by="Species", figsize=(12,8))

plt.show()
```

```python
# Plotting of multidimensional dataset in 2D
from pandas.plotting import radviz
radviz(iris_flower.drop("Id", axis=1), "Species")
```
